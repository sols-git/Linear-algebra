## Практическое задание по лекции 1 Линейная алгебра

__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

__Решение__

Заметим, что $f_{4}(x)=f_{3}(x)-f_{2}(x)-f_{1}(x)$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{3}(x)$ и $f_{2}(x)$ и $f_{1}(x)$, из чего можно сделать вывод, что $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x} $ линейно зависимы.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

__Решение__

Заметим, что $f_{4}(x)=f_{3}(x)+f_{1}(x)*f_{2}(x)+1$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{3}(x)$ и $f_{2}(x)$ и $f_{1}(x)$, из чего можно сделать вывод, что $f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2} $ линейно зависимы.

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

__Решение__

базис $b_{1}$ линейного пространства $\mathbb{R}^{3}$ образуют векторы 
$b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.
Тогда 

$$x=(2, 3, 5)= a_{1}(0,0,10)+a_{2}(2, 0, 0)+a_{3}(0, 1, 0)=(0, 0, 10a_{1})+(2a_{2}, 0, 0)+(0, a_{3}, 0)$$

используя свойства векторов, запишем:

$$x=(2, 3, 5)= 2a_{2}, a_{3}, 10a_{1} $$

находим координаты:
$$a_{1} = 0.5$$
$$a_{2} = 1$$
$$a_{3} = 3$$

__4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

$$a_{1} = 2$$

$$a_{3} = 3$$

б) $$3x^{2}-2x+2 = a_{1}x^2 +  a_{2}(x-1) + a_{3}$$

   $$-3x^{2}+2(x-1)+0 = a_{1}x^2 +  a_{2}(x-1) + a_{3}$$
используя свойства векторов, запишем:
$$-3x^{2} = a_{1}x^2$$
$$2(x-1) = a_{2}(x-1)$$
$$0 = a_{3}$$

$$a_{1} = -3$$

__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$.

__Решение__


1. Проведя проверку по приведенному утверждению, мы увидим, что 

$$(0, a, b)+(0, c, d) = (0, a+c, b+d),$$ $$\alpha\cdot (0, a, b) = (0, \alpha a, \alpha b).$$
или
$$(a, 0, b)+(c, 0, d) = (a+c, 0, b+d),$$ $$\alpha\cdot (a, 0, b) = (\alpha a, 0, \alpha b).$$
но например:
$$(a, 0, b)+(0, c, d) = (a, c, b+d),$$ 

Полученные векторы уже не принадлежат указанному в задании множеству всех векторов вида $(0, 0, b)$, то есть данное множество не является подпространством линейного пространства $\mathbb{R}^{3}$.

2. Исходя из свойства векторов, если все векторы, являются линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$, тогда данные вектора линейно зависимы. По аксиоме: _Если некоторое подмножество множества $a_{1}, a_{2},...,a_{n}$ линейно зависимо, то и все элементы $a_{1}, a_{2},...,a_{n}$ линейно зависимы_.

Соответственно, подпространство где все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$ будет тоже линейно

а) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

```python
import numpy as np
```

```python
x = np.array([0,-3,6])
y = np.array([-4,7,9])
print(f'a) Скалярное произведение x и y: {np.dot(x, y)}')
x = np.array([7,-4,0,1])
y = np.array([-3,1,11,2])
print(f'b) Скалярное произведение x и y: {np.dot(x, y)}')
```

__2.__ Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

```python
from numpy.linalg import norm
```

```python
x = np.array([4,2,4])
y = np.array([12,3,4])
print(f'норма вектора x: манхетенская  - {norm(x, ord=1)}, евклидова - {norm(x, ord=2)}')
print(f'норма вектора y: манхетенская  - {norm(y, ord=1)}, евклидова - {norm(y, ord=2)}')
cos_phi = np.dot(x, y) / norm(x) / norm(y)
print(f'Косинус угла между x и y: {cos_phi:.2f}')
print(f'Угол между x и y: {np.arccos(cos_phi):.2f}')
```
